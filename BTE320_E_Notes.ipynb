```python
email = "URGENT! Open this link: http://www.spam.org to receive your award!!!"
is_urgent = "URGENT"
has_www = "www."

if is_urgent in email:
  print("Phising!")
elif has_www in email:
  print("Spam!")
else:
  print("Legit")
```

Starts with the "if" function, if thats false it will move onto the "elif", if thats false it will move onto the "else"

```python
grade = float(input("Enter student grade (0-100): "))

if grade >= 97:
  print("A+")
elif grade >= 93:
  print("A")
elif grade >= 90:
  print("A-")
```

```python
email = "URGENT!! Click on the following: www.org to ..."
is_urgent = "URGENT"
has_www = "www."

if is_urgent in email:
  if has_www in email:
    print("Phising!")
  else:
    print("Spam!")
else:
  print("Legit!")
```

Both 5 and 6 have to be true in order to get Phising!

If only "URGENT" is correct, Spam! will show

If ALL are false Legit! will show



```python
email = "URGENT!! Click on the following: to ..."
is_urgent = "URGENT"
has_www = "www."

if (is_urgent in email) and (has_www in email):
  print("Phising!")
elif (is_urgent in email):
  print("Spam!")
else:
  print("Legit!")
```
