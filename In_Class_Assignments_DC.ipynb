```python
num_1 = float(input("Number 1: "))
num_2 = float(input("Number 2: "))
Symbol = input("Mathematical Symbol: ")

if Symbol == "+":
  print(num_1 + num_2)
elif Symbol == "-":
  print(num_1 - num_2)
elif Symbol == "/":
  print(num_1 / num_2)
elif Symbol == "*":
  print(num_1 * num_2)
elif Symbol == "**":
  print(num_1 ** num_2)
elif Symbol == "%":
  print(num_1 % num_2)
else:
  print("Invalid Operation")
```
