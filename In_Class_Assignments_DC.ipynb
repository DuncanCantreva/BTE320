```python
num_1 = float(input("Number 1: "))
num_2 = float(input("Number 2: "))
Symbol = input("Mathematical Symbol: ")

if Symbol == "+":
  print(num_1 + num_2)
elif Symbol == "-":
  print(num_1 - num_2)
elif Symbol == "/":
  print(num_1 / num_2)
elif Symbol == "*":
  print(num_1 * num_2)
elif Symbol == "**":
  print(num_1 ** num_2)
elif Symbol == "%":
  print(num_1 % num_2)
else:
  print("Invalid Operation")
```

```python
print("Hour Fee")
for p_hours in range (1,9):
  cost = (p_hours * 2.5) + 5
  if cost < 10:
    cost = 10.0
  elif cost > 20:
    cost = 20.0
  print(f"{p_hours}: ${cost} ")
```

```python
ticket = 10
fixed_costs = 200
default_attendance = 20

for ads in range (0, 201, 25):
  attendance = 2 * (ads ** .5)
  revenue = (round(attendance) + default_attendance) * ticket
  costs = fixed_costs + ads
  profit = revenue - costs
  print(f"Ad ${ads}    Profit ${profit}")
```

```python
print("Hour Fee")
p_hours = 1
while p_hours < 9:
  cost = (p_hours * 2.5) + 5
  if cost < 10:
    cost = 10.0
  elif cost > 20:
    cost = 20.0
  print(f"{p_hours}: ${cost} ")
  p_hours += 1
```

```python
ticket = 10
fixed_costs = 200
default_attendance = 20
ads = 0
while ads <= 200:
  attendance = 2 * (ads ** .5)
  revenue = (round(attendance) + default_attendance) * ticket
  costs = fixed_costs + ads
  profit = revenue - costs
  print(f"Ad ${ads}    Profit ${profit}")
  ads += 25
```

```python
start = input("Do you want to play the game? (y/n)")
import random
while start == "y":
  computer_move = random.choice(["rock", "paper", "scissors"])
  your_move = input("Your Move: ")
  print(f"Computers Move: {computer_move}")

  if your_move == computer_move:
    print("Draw")

  elif your_move == "rock" and computer_move == "scissors":
    print("You Win!")
  elif your_move == "scissors" and computer_move == "paper":
    print("You Win!")
  elif your_move == "paper" and computer_move == "rock":
    print("You Win!")
  else:
    print("You Lose!")

  start = input("Do you want to play again? (y/n)")
```

```python
names = {}

while True:
  name = input("Enter a name:")
  if name == "":
    break
  if name in names:
    names[name] += 1
  else:
    names[name] = 1
print(names)
```
